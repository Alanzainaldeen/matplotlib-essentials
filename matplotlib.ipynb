```python
fig  = plt.figure(figsize=(4,3))
ax = plt.axes()
x = np.linspace(0,20,1000)
y = np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('seaborn-v0_8-deep')
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y+y2-0.5
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
fig = plt.figure(figsize=(4,2))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig = plt.figure(figsize=(4,3))
ax = plt.axes()
ax.plot(x,y1,color='blue',linestyle='solid')
ax.plot(x,y2,color='g',linestyle='dashed')
ax.plot(x,y3,color=(1,0,0),linestyle='dotted')
```

```python
plt.style.use('seaborn-v0_8-talk')
fig = plt.figure(figsize=(6,4),dpi=100)
x1 = np.linspace(0,20,1500)
y22 = np.cos(x1)
plt.plot(x1,y22,color='orange',linestyle='dashed',linewidth=1.2)
plt.xlim(1,14)
plt.ylim(1.2,-1)
```

```python
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x,y1,'-.r',label='sin(x)')
plt.plot(x,y2,':b',label='cos(x)')
plt.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
plt.axis('equal')
plt.legend(loc='lower left')
# plt.legend(loc='upper left')
plt.title('Sin-Cos-Curves')
plt.xlabel("X values")
plt.ylabel("Y calues")
```

```python
plt.figure(figsize=(6,4),dpi=100)
ax = plt.axes()
ax.plot(x,y1,'-.r',label='sin(x)')
ax.plot(x,y2,':b',label='cos(x)')
ax.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
ax.set(xlim=(-1,11),ylim=(-5,5),xlabel='X values',ylabel='Y values',
      title='Sin-Cos-Curves')
plt.axis('equal')
plt.legend(loc='lower left')
```

```python
plt.style.use('seaborn-v0_8-talk')
fig = plt.figure(figsize=(6,4),dpi=100)
x1 = np.linspace(0,20,1500)
y22 = np.cos(x1)
ax = plt.axes()
ax.plot(x1,y22,color='orange',linestyle='dashed',linewidth=1.2,label='Cos(x)')
plt.legend(loc='upper left')
ax.set(xlim=(1,14),ylim=(-1,1.2),xlabel='X values',ylabel='Y values',
      title='Cos-Curves')
```

```python
markers = ['o','.','x','+','^','<','>','s','d']
x = np.linspace(-np.pi,np.pi,30)
y = np.cos(x)
plt.figure(figsize=(4,3),dpi=100)
plt.plot(x,y,markers[7],color='black',markersize=7)
```
